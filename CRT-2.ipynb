# Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

# What are copulas?

Copulas allow us to decompose a **joint probability distribution** into their marginals and a function which couples them together and thus allows us to specify the correlation separately.

# Gaussian Copula

**Definition:** The Gaussian Copula captures the dependence structure between random variables using the correlation matrix of a multivariate normal distribution.

A Gaussian copula with dependence parameter theta is defined here. It takes uniform random variables u1 and u2 as input and transforms them to dependent variables following a standard normal distribution.


```python
def gaussian_copula(u1, u2, theta):
  z1 = norm.ppf(u1)
  z2 = norm.ppf(u2)
  return norm.cdf(theta * z1 + np.sqrt(1 - theta**2) * z2), norm.cdf(z2)
```

```python
np.random.seed(123)  # Set random seed for reproducibility
u1 = np.random.uniform(size=1000)
u2 = np.random.uniform(size=1000)
z1, z2 = gaussian_copula(u1, u2, 0.2)
```

# Gaussian Copula Graph

```python
plt.scatter(z1, z2)
plt.xlabel("Variable 1 (Standard Normal)")
plt.ylabel("Variable 2 (Standard Normal)")
plt.title("Gaussian Copula with Theta=0.2 (Positive Dependence)")
plt.show()
```

# Explaination

This code defines a bivariate Gaussian copula, which transforms uniformly distributed variables into standard normal variables with a specific dependence structure controlled by the parameter theta. The plot shows a positive dependence between the two variables, where high values in one tend to be accompanied by high values in the other.
